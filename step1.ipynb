````python
import pandas
import sklearn.model_selection
````

# Instructions

- Read the data from CSV file and properly set the index
- Split the data into a train data set and a test data set using `sklearn.model_selection.train_test_split` with the following configurations:
```
    train_size   = 0.75
    shuffle      = True 
    random_state = 0
```
- Write the train data set to `./data/heart_failure.train.csv`
- Write the test data set to `./data/heart_failure.test.csv`
- **No data processing shall be performed in this step**


````python
df = pandas.read_csv('./data/high_salary.train.csv').set_index("id")
````

````python
df.head()
````

````python
data_train, data_test = sklearn.model_selection.train_test_split(
    df,
    train_size   = 0.75,
    shuffle      = True,
    random_state = 0,
)
````
